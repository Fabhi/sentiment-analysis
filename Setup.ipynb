```python
import nltk
import random

from nltk.classify.scikitlearn import SklearnClassifier
from nltk.tokenize import word_tokenize

import pickle
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC, LinearSVC, NuSVC
from nltk.classify import ClassifierI
from statistics import mode
from nltk.tokenize import word_tokenize
import re
import csv

from nltk.corpus import stopwords
```

```python
bag_of_words = []
labelled_tweets = []

labels = ["negative", "positive"]

stopwords = list(set(stopwords.words('english')))
```

```python
train_file = open('train_tweets.csv', encoding="utf-8")
reader = csv.reader(train_file, delimiter=',')
for index, datatuple in enumerate(reader):
    #print(datatuple, index)
    if index != 0:
        labelled_tweets.append((datatuple[2], labels[int(datatuple[1])]))

        sanitized_words = re.sub(r'[^(a-zA-Z)\s]', '', datatuple[2])
        tokens = word_tokenize(sanitized_words)
        tokens = [word for word in tokens if not word in stopwords]

        part_of_speech = nltk.pos_tag(tokens)

        for word in part_of_speech:
            if "J" in word[1]:
                bag_of_words.append(word[0].lower())
                

    
    
```

```python
## Feature Sets
```

```python
bag_of_words = nltk.FreqDist(bag_of_words)
features = list(bag_of_words.keys())[:5000]
# print(features)
```

```python
def featurise(tweet):
    word_tokens = word_tokenize(tweet)
    features_map = {}
    for word in features:
        features_map[word] = (word in word_tokens)
    return features_map
```

```python
feature_set = []
for (tweet, sentiment) in labelled_tweets:
    #print(tweet)
    feature_set.append((featurise(tweet), sentiment))
print(feature_set[:2])
```

```python
random.shuffle(feature_set)
```

```python
training_features = feature_set[:15000]
testing_features = feature_set[15000:]
```

```python
# Training using the feature set
```

```python
NBClassifier = nltk.NaiveBayesClassifier.train(training_features)
```

```python
print("Accuracy:", nltk.classify.accuracy(NBClassifier, testing_features))
```
